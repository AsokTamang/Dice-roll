```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_dice = 6
dice =np.array([i for i in range(1,n_dice + 1)])
dice  
```

```python
n_rolls = 20
output = np.array([np.random.choice(dice) for i in range(n_rolls)])  #here we are storing the output of dice roll upto given n times
output
```

```python
m = np.mean(output)  #mean of the output
v = np.var(output)  #variance of the output
print(f'Mean of the output is :{m:.2f}\nVariance of the output is :{m:.2f} of the roll dice')
```

```python
sns.histplot(output,discrete=True)  #showing the count of each number in each roll of the dice output
```

```python
#now when the two dice is rolled 
time_of_rolls = 20000
first_rolls = np.array([np.random.choice(dice) for i in range(time_of_rolls)])
second_rolls = np.array([np.random.choice(dice) for i in range(time_of_rolls)])
sum_of_rolls = first_rolls + second_rolls
sum_of_rolls
```

```python
covariance = np.cov(first_rolls,second_rolls)
print(f'Covariance between first and second roll is:', covariance)
#as we can see that the covariance between the variables of first roll and the second roll is nearly equal to 0, cause the first roll and the second roll are independent of eachother
```

```python
sns.histplot(sum_of_rolls,stat="probability",discrete=True)
#as we can see the histogram of the probability distribution of sum of rolls is a normal distribution
```

```python
fair_probability = np.array([1/n_dice] *n_dice)  #this is the probability of fair dice
sns.barplot(x=dice,y=fair_probability)
plt.show()  #as we can observe the fair dice has the probability same for all the face of a dice
```

```python
def loaded_dice(n_size,choosen_number):  #here choosen number means the number which has the high probability of being rolled
    probability_values = np.array([(1/(n_size + 1)) for i in range(n_size)])
    probability_values[choosen_number - 1] = 1 - sum(probability_values[:-1])  #here we are assigning the higher probability to this given choosen number
    if not np.isclose(sum(probability_values),1):
        print('The sum of probability of all the variables must be 1.')
        return
    return probability_values
assigned_probabilities =loaded_dice(6,2)    #so we have assigned number 2 to roll more or it has higher probability to appear in dice roll compared to other numbers
sns.barplot(x=dice,y=assigned_probabilities)
#as we can see from the barchart below the probability of the number 2 is comparatively higher compared to other numbers
```

```python
#now checking with much higher number of rolls and assigned probability
higher_rolls = 20000
first_rolls = np.array([np.random.choice(dice,p=loaded_dice(6,6)) for i in range(higher_rolls)])  #here we are making the number 6 priority
second_rolls = np.array([np.random.choice(dice,p=loaded_dice(6,3)) for i in range(higher_rolls)])  #here we are making 3 priority
final_sum_rolls = first_rolls + second_rolls
sns.histplot(final_sum_rolls,stat='probability',discrete=True)
```
