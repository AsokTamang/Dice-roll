```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_dice = 6
dice =np.array([i for i in range(1,n_dice + 1)])
dice  
```

```python
n_rolls = 20
output = np.array([np.random.choice(dice) for i in range(n_rolls)])  #here we are storing the output of dice roll upto given n times
output
```

```python
m = np.mean(output)  #mean of the output
v = np.var(output)  #variance of the output
print(f'Mean of the output is :{m:.2f}\nVariance of the output is :{m:.2f} of the roll dice')
```
