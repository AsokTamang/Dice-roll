```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iterations = 20000
dice = np.array([1,2,3,4,5,6])
ans = []
count = 0
for i in range(iterations):
 first_roll = np.random.choice(dice) 
 second_roll = np.random.choice(dice)
 sum_roll = first_roll + second_roll
 ans.append(sum_roll)
for value in ans:
    if value > 5:
        count+=1
print('The probability that the sum of two throws is greater than 5 is', count/len(ans))

        
```
